**Scenarios where each scaling technique is preferred and why?**
Standard Scaling:
- Used when data should be centered at mean 0 and variance 1.
- Works best for algorithms assuming Gaussian distribution:
  (Logistic Regression, Linear Regression, SVM, PCA)

Min-Max Scaling:
- Compresses values to range 0â€“1.
- Useful for distance-based models:
  (KNN, Neural Networks, Clustering)

```python
cat_cols = df.select_dtypes(include = ['object']).columns
cat_cols
```

```python
for col in cat_cols:
  print(col, ":", df[col].nunique())
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
one_hot_cols = []
label_cols = []

for col in cat_cols:
  if df[col].nunique() <5:
    one_hot_cols.append(col)
  else:
    label_cols.append(col)

  df_encoded = pd.get_dummies(df, columns = one_hot_cols, drop_first = True)#drop_first=True kept the column count the same.

  le = LabelEncoder()
  for col in label_cols:
    df_encoded[col] = le.fit_transform(df[col])
df_encoded.head(10)

```

```python
print("\nOriginal Shape:", df.shape)
print("Encoded Shape:", df_encoded.shape)
```

```python
df['hours_per_week'] = pd.cut(df['hours_per_week'], bins = 3, labels = ['low', 'medium', 'high'])
df.head(10)
```

```python
df['age_group'] = pd.cut(df['age'], bins=3, labels=['Young', 'Adult', 'Senior'])
df.head(10)
```

```python
df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

```python
print("Original Skewness:", df['capital_gain'].skew())
print("Log Skewness:", df['capital_gain_log'].skew())
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(n_estimators=200, contamination=0.02, random_state=42)

# Fit on numeric features
numeric_cols = df_encoded.select_dtypes(include=['int64','float64']).columns
iso_pred = iso.fit_predict(df_encoded[numeric_cols])

# Mark outliers
df_encoded['is_outlier'] = (iso_pred == -1).astype(int)

# Remove outliers
df_no_outliers = df_encoded[df_encoded['is_outlier'] == 0].drop(columns=['is_outlier'])
```

```python
import ppscore as pps

pps_matrix = pps.predictors(df_no_outliers, 'income')
pps_matrix.sort_values(by='ppscore', ascending=False).head(10)
```
